```r
# Binomial Distribution in R
n <- 10  # Number of trials
p <- 0.5 # Probability of success

# Probability Mass Function (PMF) for Binomial Distribution
x <- 0:n
probabilities <- dbinom(x, size=n, prob=p)

# Plotting the Binomial Distribution
barplot(probabilities, names.arg=x, col="blue",
        main="Binomial Distribution (n=10, p=0.5)",
        xlab="Number of Successes", ylab="Probability")
```

```r
# Poisson Distribution in R
lambda <- 4  # Average number of occurrences

# Probability Mass Function (PMF) for Poisson Distribution
x <- 0:15
probabilities <- dpois(x, lambda=lambda)

# Plotting the Poisson Distribution
barplot(probabilities, names.arg=x, col="red",
        main="Poisson Distribution (λ=4)",
        xlab="Number of Occurrences", ylab="Probability")
```

```r
# Normal Distribution in R
mean <- 50    # Mean
sd <- 10      # Standard Deviation

# Generating a sequence of values
x <- seq(mean - 4*sd, mean + 4*sd, length=100)
density <- dnorm(x, mean=mean, sd=sd)

# Plotting the Normal Distribution
plot(x, density, type="l", col="green", lwd=2,
     main="Normal Distribution (mean=50, sd=10)",
     xlab="Values", ylab="Density")
```
